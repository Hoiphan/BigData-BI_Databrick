```python
dbutils.widgets.removeAll()
dbutils.widgets.text("weight_kg", "", "Weight (kg)")
dbutils.widgets.text("height_m", "", "Height (meters, e.g. 1.75)")
dbutils.widgets.dropdown("sex", "ALL", ["M", "F", "ALL"], "Sex")
dbutils.widgets.text("start_year", "", "Start year (YYYY)")
dbutils.widgets.text("start_month", "", "Start month (MM)")
dbutils.widgets.text("start_day", "", "Start day (DD)")
dbutils.widgets.text("end_year", "", "End year (YYYY)")
dbutils.widgets.text("end_month", "", "End month (MM)")
dbutils.widgets.text("end_day", "", "End day (DD)")
dbutils.widgets.text("time_range", "", "Time range (HHMMSSsss,HHMMSSsss as int)")

# Cast weight and height input to float
weight_kg = float(dbutils.widgets.get("weight_kg")) if dbutils.widgets.get("weight_kg") else None
height_m = float(dbutils.widgets.get("height_m")) if dbutils.widgets.get("height_m") else None
```

```python
%sql
SELECT COUNT(*) AS study_count
FROM dbms.gold.dim_study
WHERE (
    ((:start_year = '' OR :start_month = '' OR :start_day = ''))
    OR (
        :start_year <> '' AND :start_month <> '' AND :start_day <> ''
        AND study_date >= try_cast(concat(:start_year, '-', lpad(:start_month, 2, '0'), '-', lpad(:start_day, 2, '0')) AS DATE)
    )
)
AND (
    ((:end_year = '' OR :end_month = '' OR :end_day = ''))
    OR (
        :end_year <> '' AND :end_month <> '' AND :end_day <> ''
        AND study_date <= try_cast(concat(:end_year, '-', lpad(:end_month, 2, '0'), '-', lpad(:end_day, 2, '0')) AS DATE)
    )
)
AND (:sex = 'ALL' OR patient_sex = :sex)
AND (:height_m = '' OR patient_height = try_cast(:height_m AS DOUBLE))
AND (:weight_kg = '' OR patient_weight = try_cast(:weight_kg AS INT))
AND (:time_range = '' OR study_time = try_cast(:time_range AS INT))
;
```
